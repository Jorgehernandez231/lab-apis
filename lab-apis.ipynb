```python
artists = ["Lia Kali", "Leiva", "Estopa"]

def get_top_tracks(artist_name):

    # 3a) Search for the artist to get their Spotify ID
    results = sp.search(q=f"artist:{artist_name}", type="artist", limit=1)
    items = results.get("artists", {}).get("items", [])
    if not items:
        return []  # no such artist found
    
    artist_id = items[0]["id"]
    
    # 3b) Get the artist’s top tracks
    top_tracks_data = sp.artist_top_tracks(artist_id)
    tracks = top_tracks_data.get("tracks", [])
    
    # 3c) Return the first 5 track names
    return [track["name"] for track in tracks[:5]]

# 4) Build a list of each artist’s top 5 tracks
top_tracks_list = [get_top_tracks(artist) for artist in artists]

# 5) Inspect the result
for artist, top_tracks in zip(artists, top_tracks_list):
    print(f"{artist}:")
    for i, title in enumerate(top_tracks, start=1):
        print(f"  {i}. {title}")
    print()
    
   
```

```python
from spotipy.exceptions import SpotifyException
import requests
```

```python
artists = ["Lia Kali", "Leiva", "Estopa"]

def get_top_tracks(artist_name):
    results = sp.search(q=f"artist:{artist_name}", type="artist", limit=1)
    artist_items = results['artists']['items']
    if not artist_items:
        return []
    artist_id = artist_items[0]['id']
    
    top_tracks_data = sp.artist_top_tracks(artist_id, country='ES')  # Use your preferred market
    top_tracks = [track['name'] for track in top_tracks_data['tracks'][:5]]
    return top_tracks

top_tracks_list = []
for artist in artists:
    top_tracks = get_top_tracks(artist)
    top_tracks_list.append({artist: top_tracks})

# 4. Discover related artists


def find_related_artists(artist_name):
    try:
        results = sp.search(q=f"artist:{artist_name}", type="artist", limit=1)
        artist_items = results['artists']['items']
        if not artist_items:
            print(f"❌ Artist not found: {artist_name}")
            return []

        artist_id = artist_items[0]['id']

        try:
            related = sp.artist_related_artists(artist_id)
            related_names = [artist['name'] for artist in related['artists'][:5]]
            return related_names

        except (SpotifyException, requests.exceptions.HTTPError) as e:
            print(f"⚠️ No related artists found for '{artist_name}': {e}")
            return []

    except Exception as e:
        print(f"❗ Unexpected error for '{artist_name}': {e}")
        return []


related_artists_list = []
for artist in artists:
    related = find_related_artists(artist)
    related_artists_list.append({artist: related})

# 5. 🎧 Challenge: Create a playlist of top tracks from all related artists
playlist = []

# Top tracks of original artists
for entry in top_tracks_list:
    for artist, tracks in entry.items():
        playlist.extend(tracks)

# Top tracks of related artists
for entry in related_artists_list:
    for _, related_artists in entry.items():
        for related_artist in related_artists:
            playlist.extend(get_top_tracks(related_artist))

# Optional: Remove duplicates
playlist = list(dict.fromkeys(playlist))

# Display the result
print("\n🎶 Your Discover Playlist:")
for idx, track in enumerate(playlist, 1):
    print(f"{idx}. {track}")
```

If you want to get the tracks from a specific playlist, you can use the playlist_tracks method.

```python
tracks = sp.playlist_tracks(playlist_id)
for track in tracks['items']:
    print(track['track']['name'])  # Print each track's name
```

### Getting Artists from a Playlist

To extract all the artists from the tracks in a playlist, you'd typically follow these steps:

1. Fetch the playlist's tracks.
2. Iterate through each track.
3. For each track, extract the associated artists.

```python
# List to store unique artist names
artists_list = []

for track_item in tracks['items']:
    track = track_item['track']
    for artist in track['artists']:
        artist_name = artist['name']
        if artist_name not in artists_list:  # This ensures each artist is added only once
            artists_list.append(artist_name)

print(artists_list)
```

## Exercise 2: Unraveling the World of Playlists


1. **Featured Exploration**: 
   - Fetch the list of Spotify's current featured playlists. 
   - Extract and display the names and IDs of the top 5 featured playlists.
   
2. **Deep Dive**:
   - Choose any one of the top 5 featured playlists (you can choose the one you personally find most interesting or simply pick one randomly).
   - Fetch and display its name, description, and total track count.

3. **Track-tastic**:
   - Extract and display the names of the first 10 tracks in the chosen playlist.

4. **Artistic Flair**:
   - Create a dictionary where the keys are the names of the first 10 tracks, and the values are lists containing the names of the artists associated with each track.
   - For example: `{"TrackName1": ["Artist1", "Artist2"], "TrackName2": ["Artist3"]}`
   

```python
from dotenv import load_dotenv
load_dotenv()   # this reads .env and sets os.environ

from spotipy.oauth2 import SpotifyClientCredentials
import spotipy

auth_manager = SpotifyClientCredentials()
sp = spotipy.Spotify(auth_manager=auth_manager)
```

```python
featured = sp.featured_playlists(limit=5)                # get top 5
playlists = featured["playlists"]["items"]

print("Top 5 Featured Playlists:")
for i, pl in enumerate(playlists, start=1):
    print(f"{i}. {pl['name']}  (ID: {pl['id']})")


# 2. Deep Dive
# ───────────────────────────────────────────────────────────────────────────────
# Let's pick the first one:
chosen = playlists[0]
pid    = chosen["id"]

details = sp.playlist(pid, fields="name,description,tracks.total")
print("\nDeep Dive — Playlist Details:")
print("Name:       ", details["name"])
print("Description:", details["description"] or "(no description)")
print("Total Tracks:", details["tracks"]["total"])


# 3. Track-tastic
# ───────────────────────────────────────────────────────────────────────────────
first_10 = sp.playlist_items(pid, limit=10)["items"]
track_names = [item["track"]["name"] for item in first_10]

print("\nFirst 10 Tracks:")
for idx, name in enumerate(track_names, start=1):
    print(f"{idx}. {name}")


# 4. Artistic Flair
# ───────────────────────────────────────────────────────────────────────────────
track_to_artists = {
    item["track"]["name"]: [artist["name"] for artist in item["track"]["artists"]]
    for item in first_10
}

print("\nTrack → Artists Mapping:")
for track, artists in track_to_artists.items():
    print(f"• {track}: {artists}")
```
